### Exercise 1

Create a 1-D Numpy Array including at least 8 int64 numbers. Show the matrix dimensions and shape. 

```python
import numpy as np

# Create the array

arr = np.array([12, 64, 101, 44, 78, 65, 25, 2])

print(arr)

print("Number of dimensions: " + str(arr.ndim))

print("Matrix shape: " + str(arr.shape))
```

### Exercise 2

With the matrix from exercise 1, calculate the average and substract that number from each value in the array.


```python
# Calculate average:

mitjana = sum(arr) / len(arr)

print("The average is " + str(mitjana))

# Substract average:

for i in range(len(arr)):
    arr[i] -= mitjana

print(arr)
```

### Exercise 3

Create a 2-D matrix with 5x5 shape. Extract the maximum value of the matrix and the maximum value of each of their axes.


```python
# Create matrix

matrix = np.array([
    [12, 30, 59, 21, 55],
    [100, 59, 37, 98, 1],
    [10, 55, 99, 12, 49],
    [28, 57, 69, 92, 87],
    [36, 81, 48, 12, 56],
])

# Absolute maximum value:

print("The maximum value overall is: " + str(np.max(matrix)))

# Maximum values in each level:

for i in range(len(matrix)):     
    value = str(max(matrix[i]))
    print(f"Maximum value in row {i}: {value}")
```

```python
# Create matrixes:

exempleBroadcast = np.array([
    [12, 30, 59, 21, 55],
    [100, 59, 37, 98, 1],
    [12, 30, 59, 21, 55]
])

second = np.array([
    [10, 55, 99, 12, 49],
    [28, 57, 69, 92, 87],
    [36, 81, 48, 12, 56],
])

# Operate between them:

result = exempleBroadcast + second

resta = exempleBroadcast - second

print("Adding up the two matrixes:")

print(result)

print("Substracting the matrixes:")

print(resta)

just1 = exempleBroadcast - 20

print("Substracting a single value from all matrix elements:")

print(just1)

different = np.array([
    [10, 55, 99, 12, 49],
    [28, 57, 69, 92, 87]
])

print("Evidence that incompatible dimensions cannot be operated:")

problem = exempleBroadcast - different
```
